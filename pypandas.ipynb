```python
# Q2
import xml.etree.ElementTree as ET
purchase_orders = ET.parse('task3.xml')

root = purchase_orders.getroot()
company = root.get('company')
print('The name of the company is', company)
#root.attrib.values()
```

```python
# Q3.a
from datetime import datetime
date_format = "%m/%d/%y"
my_list=[]
for order in root:
        order_date = order.find('OrderDate')
        arrival_date = order.find('ArrivalDate')
        b=datetime.strptime(order_date.text,date_format)
        c=datetime.strptime(arrival_date.text,date_format)
        delivery_time=c-b
        days=delivery_time.days
        my_list.append(days)

print('The minimum delivery time of purchase orders is',min(my_list),'days.')
print('The maximum delivery time of purchase orders is',max(my_list),'days.')
print('The average delivery time of purchase orders is',round(sum(my_list)/len(my_list),2),'days.')
```

Q3.a.The minimum delivery time of purchase orders is 3 days, the maximum delivery time of purchase orders is 30 days, and the average delivery time of purchase orders is 8.41 days.

```python
#Q3.b
short_time=[s_element for s_element in my_list if s_element < 5]
short_time_count = len(short_time)
medium_time=[m_element for m_element in my_list if m_element >= 5 and m_element <= 9]
medium_time_count = len(medium_time)
long_time=[l_element for l_element in my_list if l_element > 9]
long_time_count=len(long_time)
print(short_time_count,'orders are classified as short.')
print(medium_time_count,'orders are classified as medium.')
print(long_time_count,'orders are classified as long.')    
```

Q3.b. There are 3 orders are classified as short, 68 orders are classified as medium, 23 orders are classified as long.

```python
#Q4
import xml.etree.ElementTree as ET
from lxml import etree
import pandas as pd

my_list_str = [str(x) for x in my_list]
my_list_class = my_list.copy()
for i, n in enumerate(my_list_class):
    if n < 5:
        my_list_class[i] = "short"
    elif n >= 5 and n <= 9:
        my_list_class[i] = "medium"
    else:
        my_list_class[i] = "long"

purchase_orders = ET.parse('task3.xml')
root = purchase_orders.getroot()
child = ET.SubElement(root,"Order")
xmlstr = ET.tostring(root, encoding='utf8',method='xml')
doc = etree.fromstring(xmlstr)
for arrivaldate in doc.xpath('/PurchaseOrders/Order/ArrivalDate[1]'):
    parent = arrivaldate.getparent()
    parent.insert(parent.index(arrivaldate)+1, etree.XML('<DeliveryTime> </DeliveryTime>\n'))
    parent.insert(parent.index(arrivaldate)+2, 
                  etree.XML('<DeliveryTimeClassification> </DeliveryTimeClassification>\n'))
for i in range(0,94):
    delivery_time_index = doc[i][10]
    delivery_time_index.text = my_list_str[i]+' days'
    delivery_time_class_index = doc[i][11]
    delivery_time_class_index.text = my_list_class[i]
    i+=1
new_doc = etree.tostring(doc)
tree_final = ET.ElementTree(doc)    
df_cols = ["Supplier","OrderNo","ItemNo","ItemDescription","ItemCost","Quantity","CostPerOrder","APTerms","OrderDate","ArrivalDate","DeliveryTime","DeliveryTimeClassification"]
rows = []
for n1 in range(0, 94):
    sup = doc[n1][0].text
    orderno = doc[n1][1].text
    itemno = doc[n1][2].text
    itemdesc = doc[n1][3].text
    itemcost = doc[n1][4].text
    quantity = doc[n1][5].text
    cpo = doc[n1][6].text
    art = doc[n1][7].text
    orderdt = doc[n1][8].text
    arrdt = doc[n1][9].text
    delivtime = doc[n1][10].text
    delivclass = doc[n1][11].text
    rows.append({"Supplier":sup, "OrderNo":orderno, "ItemNo":itemno,"ItemDescription":itemdesc,"ItemCost":itemcost,"Quantity":quantity,"CostPerOrder":cpo,"APTerms":art,"OrderDate":orderdt,"ArrivalDate":arrdt,"DeliveryTime":delivtime,"DeliveryTimeClassification":delivclass})
out_df = pd.DataFrame(rows, columns = df_cols)
out_df.to_csv('Task3_745973306.csv')
out_df
```
